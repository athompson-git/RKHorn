# Flux Plotting for RKHorn

### Energy Distributions

```python
import numpy as np
import matplotlib.pyplot as plt

mb_target_pot_nu = 1.875e21
mb_target_pot_nubar = 1.127e21

# RKHorn beam data
pip_flux_mb = np.genfromtxt("data/bnb_focused_piplus_flux_rkhorn_1POT_BetheBloch.txt")
pip_flux_mb[:,-1] *= mb_target_pot_nu

pim_flux_mb = np.genfromtxt("data/bnb_focused_piminus_flux_rkhorn_1POT_BetheBloch.txt")
pim_flux_mb[:,-1] *= mb_target_pot_nubar

pip_energies = pip_flux_mb[:,0]
pip_angles = pip_flux_mb[:,1]
pip_wgts = pip_flux_mb[:,-1]
pip_z = pip_flux_mb[:,4]

pim_energies = pim_flux_mb[:,0]
pim_angles = pim_flux_mb[:,1]
pim_wgts = pim_flux_mb[:,-1]
pim_z = pim_flux_mb[:,4]
```

```python
z_bins = np.logspace(-1, 2, 100)
plt.hist(pip_z, bins=z_bins)
plt.xscale('log')
```

```python
energy_bins = np.linspace(0.0, 5000.0, 100)

print("No. pi+ = {}".format(np.sum(pip_wgts)/mb_target_pot_nu))
print("No. pi- = {}".format(np.sum(pim_wgts)/mb_target_pot_nubar))

plt.hist(pip_energies, weights=pip_wgts*(pip_z > 2.0), bins=energy_bins, histtype='step', label=r"$\pi^+$")
plt.hist(pim_energies, weights=pim_wgts*(pim_z > 2.0), bins=energy_bins, histtype='step', label=r"$\pi^-$")
plt.yscale('log')
plt.xlim((energy_bins[0], energy_bins[-1]))
plt.show()
```

```python
plt.hist(pip_energies, weights=pip_wgts, bins=energy_bins, histtype='step', label=r"$\pi^+$", density=True)
plt.hist(pim_energies, weights=pim_wgts, bins=energy_bins, histtype='step', label=r"$\pi^-$", density=True)
plt.yscale('log')
plt.xlim((energy_bins[0], energy_bins[-1]))
plt.show()
```
